Now, Performing the Standardizing Zero Mean

```python
#importing the scipy. stats module specializes in random variables and probability distributions. 
from scipy import stats
```

```python
#importing the Matplot library
#Matplotlib makes easy things easy and hard things possible.
import matplotlib.pyplot as plt
```

```python
ad_traingDF['scaled_age'] = stats.zscore(ad_traingDF.age, nan_policy='omit')
ad_traingDF.age.plot.density(color="purple")
plt.title('plot for scaled Age')
plt.show()
```

```python
ad_traingDF['unscaled_age'] = stats.zscore(ad_traingDF.age, nan_policy='omit')
ad_traingDF.age.plot.density(color="navy")
plt.title('plot for Unscaled Age')
plt.show()
```
