2.Creating a contingency table of workclass and sex, and ploting it.

7.Standardize (zero mean) the age variable, and identify how many outliers there are. What is the most extreme outlier?

Now, Performing the Standardizing Zero Mean

```python
#importing the scipy. stats module specializes in random variables and probability distributions. 
from scipy import stats
```

```python
#importing the Matplot library
#Matplotlib makes easy things easy and hard things possible.
import matplotlib.pyplot as plt
```

```python
ad_traingDF['scaled_age'] = stats.zscore(ad_traingDF.age, nan_policy='omit')
ad_traingDF.age.plot.density(color="purple")
plt.title('plot for scaled Age')
plt.show()
```

```python
ad_traingDF['unscaled_age'] = stats.zscore(ad_traingDF.age, nan_policy='omit')
ad_traingDF.age.plot.density(color="navy")
plt.title('plot for Unscaled Age')
plt.show()
```

Describing the scaled age

```python
ad_traingDF['scaled_age'].describe()
```

```python
#plot for scaled Age outliers
adult_trainingDF_outliers = ad_traingDF.query('(scaled_age > 3 | scaled_age <-3)')['scaled_age']

```

```python
#showing the plot for Scaled Age Outliers
adult_trainingDF_outliers.plot.density()
plt.title('Plot for Scaled Age Outliers')
plt.show()
```
