**2. Create a contingency table of workclass and sex, and plot it.**

```r
# plotting barplot for sex and workclass
barplot_sex_workclass <- table(ad_trainigDF$sex,ad_trainigDF$workclass)
colors <- c("red", "green")
barplot(barplot_sex_workclass, col = colors, beside = TRUE)
legend("topright", legend = c("Female", "Male"), fill = colors)
```

**6. Determine whether any outliers exist for the education field.**

```r
boxplot(ad_trainigDFOver40$education,ylab = "education",main = "Boxplot")
```

```r
# finding upper and lower bound of age
q1 <- quantile(ad_trainigDF$education,0.25)
q3 <- quantile(ad_trainigDF$education,0.75)
iqr <- q3-q1
lower_bound <- ad_trainigDF$education[ad_trainigDF$education<(q1-(1.5 * iqr))]
upper_bound <- ad_trainigDF$education[ad_trainigDF$education>(q3+(1.5 * iqr))]
```

```r
length(lower_bound)
```

```r
length(upper_bound)
```

Total outliers are 552 at lower bound and there is no outlier at upper bound.

**7. Standardize (zero mean) the age variable, and identify how many outliers there are. What is the most extreme outlier?**

```r
boxplot(ad_trainigDF$age,ylab = "age",main = "Boxplot")
```

```r
# finding upper and lower bound of age
q1 <- quantile(ad_trainigDF$age,0.25)
q3 <- quantile(ad_trainigDF$age,0.75)
iqr <- q3-q1
lower_bound2 <- ad_trainigDF$age[ad_trainigDF$age<(q1-(1.5 * iqr))]
upper_bound2 <- ad_trainigDF$age[ad_trainigDF$age>(q3+(1.5 * iqr))]
length(lower_bound2)
length(upper_bound2)
```
